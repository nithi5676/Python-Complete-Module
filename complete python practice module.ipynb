```python
# to find the factorial of n values 

def factorial(x):
  fact=1
  for i in range(1,x+1):
    fact=fact*i
  return fact

factorial(5)    
```

# **Overridding And Inheritance**

```python
class Shape :
    
    
    def set_color(self, color):
        self.color = color
        
    def calculate_area(self):
        pass
        
    def color_the_shape(self):
        color_price = {"red" : 10, "blue" : 15, "green" : 5}
        return self.calculate_area() * color_price[self.color]
```

```python
class Circle(Shape) :
    pi = 3.14
    def __init__(self, radius):
        self.radius = radius
        
    def calculate_area(self):
        return Circle.pi * self.radius
```

```python
c = Circle(67)
c.set_color("red")
print("Circle with radius =",c.radius ,"when colored", c.color,"costs $",c.color_the_shape())
```

```python
class A:
    def __init__(self, x=1):
        self.x = x
class B(A):
    def __init__(self,y =2):
        super().__init__()
        self.y = y

def main():
    b = B()
    print(b.x,b.y)
main()
 
```
